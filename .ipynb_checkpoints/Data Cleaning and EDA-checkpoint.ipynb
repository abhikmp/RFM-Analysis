### PARETO PRINCIPLE

PARETO PRINCIPLE STATES :
1. 80% of companies revenue come from 20% of total customers.
2. 20% of products are responsible for 80% of total revenue
3. 20% of products contribute to 80% of sales volume.

#### 80% of company's revenue comes from 20% customers

```python
sales_per_customer = dft.groupby(['CustomerID'])['TotalSales'].sum().sort_values(ascending=False).reset_index()
# sales_per_customer

sales_per_customer['CummPercentage'] = (sales_per_customer['TotalSales'].cumsum()*100)/sales_per_customer['TotalSales'].sum()
# sales_per_customer

# checking if pareto principle holds true
len(sales_per_customer.query('CummPercentage < 80'))*100/len(sales_per_customer)
```

#### 20% of products are responsible for 80% of total revenue

```python
sales_by_product = dft.groupby(['StockCode'])['TotalSales'].sum().sort_values(ascending=False).reset_index()
sales_by_product

sales_by_product['CummPercent'] = (sales_by_product['TotalSales'].cumsum()*100)/sales_by_product['TotalSales'].sum()
sales_by_product

(len(sales_by_product.query('CummPercent < 80'))*100)/len(sales_by_product)
```

#### 20% of products contribute to 80% of sales volumns

```python
volume_per_product = dft.groupby(['StockCode'])['Quantity'].sum().sort_values(ascending=False).reset_index()
volume_per_product

volume_per_product['CummPercent'] = (volume_per_product['Quantity'].cumsum()*100)/volume_per_product['Quantity'].sum()
# volume_per_product

(len(volume_per_product.query('CummPercent < 80'))*100)/len(volume_per_product)
```

# RFM ANALYSIS

```python
df.info()
```

### Creating RFM Table

```python
# for recency caluclation, we'll consider the day after the max date as analysis date/latest_date
analysis_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)
analysis_date
```

```python
df_rfm = df.groupby(['CustomerID']).agg(
    Recency = ('InvoiceDate', lambda x: (analysis_date - x.max()).days),
    Frequency = ('InvoiceNo', 'nunique'),
    Monetory = ('TotalSales', 'sum')
).reset_index()

df_rfm
```

```python
df_rfm.describe()
```

### Assigning RFM Scores out of 5

```python
# Ranking Frequency
df_rfm['F_Rank'] = df_rfm['Frequency'].rank(method='first')

# get bin edges for R, F, M
_, freq_bins = pd.qcut(df_rfm['F_Rank'], q=5, retbins=True, labels=[1, 2, 3, 4, 5])
_, rec_bins = pd.qcut(df_rfm['Recency'], q=5, labels=[5, 4, 3,2,1], retbins=True)
_, mon_bins = pd.qcut(df_rfm['Monetory'], q=5, labels=[5, 4, 3,2,1], retbins=True)

actual_freq_bins = []
for i in range(len(freq_bins) - 1):
    lower_rank = freq_bins[i]
    upper_rank = freq_bins[i + 1]
    actual_lower = df_rfm.loc[df_rfm["F_Rank"] >= lower_rank, "Frequency"].min()
    actual_upper = df_rfm.loc[df_rfm["F_Rank"] <= upper_rank, "Frequency"].max()
    actual_freq_bins.append(f"{actual_lower} - {actual_upper}")

# Create a DataFrame with readable bin ranges
rfm_bin_ranges1 = pd.DataFrame({
    "Score": [1,2,3,4,5],
    "Frequency Range (Mapped)": actual_freq_bins,
    "Monetary Range": [f"{int(mon_bins[i])} - {int(mon_bins[i+1])}" for i in range(len(mon_bins)-1)]
})
rfm_bin_ranges2 = pd.DataFrame({
    "Score": [5,4,3,2,1],
    "Recency Range": [f"{int(rec_bins[i])} - {int(rec_bins[i+1])}" for i in range(len(rec_bins)-1)]
})

# Display bin ranges
print(rfm_bin_ranges1)
print(rfm_bin_ranges2)
print(actual_freq_bins)
```

```python
df_rfm['RScore'] = pd.qcut(df_rfm['Recency'], q=5, labels=[5, 4, 3, 2, 1]) # lower is better, so reversing the score
df_rfm['FScore'] = pd.qcut(df_rfm['F_Rank'], q=5, labels=[1, 2, 3, 4, 5]) # higher is better
df_rfm['MScore'] = pd.qcut(df_rfm['Monetory'], q=5, labels=[1, 2, 3, 4, 5]) # highes is better

df_rfm.head()
```

### Segmenting Customers Based on RFM Scores

Here are the segments
1. R and F == 5: Highe value
2. R and F == 4: loyal
3. R and F >= 2: at risk
4. R and F < 2: Dormant

```python
df_rfm['RScore'] = df_rfm['RScore'].astype(int)
df_rfm['FScore'] = df_rfm['FScore'].astype(int)
df_rfm['MScore'] = df_rfm['MScore'].astype(int)
```

```python
def rfm_segment(row):
    if row["RScore"] == 5 and row["FScore"] == 5: 
        return "High Value"   
    if row["RScore"] >= 4 and row["FScore"] >= 4: 
        return "Loyal"
    elif row["RScore"] >= 2 or row["FScore"] >= 2: 
        return "At-Risk"
    else:
        return "Dormant"
```

```python
df_rfm['Customer_Segment'] = df_rfm.apply(rfm_segment, axis=1)
```

```python
df_rfm['Customer_Segment'].value_counts()
```

```python
df_rfm
```

```python
# creating heatmaps
rfm_pivot = df_rfm.pivot_table(index='RScore', columns='FScore', values='MScore', aggfunc='mean', fill_value=0)
rfm_pivot

plt.figure(figsize=(10, 6))
sns.heatmap(rfm_pivot, cmap='coolwarm', annot=True, fmt='.2f', linewidths=0.5)

plt.xlabel("F_Score (Frequency Score)")
plt.ylabel("R_Score (Recency Score)")
plt.title("M-Score Heatmap (Avg Spending per RScore-FScore)", fontsize=12, fontweight='bold')

# saving file:
plt.savefig("graph1.png", dpi=300, bbox_inches='tight', transparent=True)
plt.show()
```

```python
rfm_pivot = df_rfm.pivot_table(index='RScore', columns='FScore', values='MScore', aggfunc='count', fill_value=0)

plt.figure(figsize=(10, 6))
sns.heatmap(rfm_pivot, cmap='coolwarm', annot=True, fmt='.2f', linewidths=0.5)

plt.xlabel("F_Score (Frequency Score)")
plt.ylabel("R_Score (Recency Score)")
plt.title("M-Score Heatmap (Avg Spending Per RScore-FScore)", fontsize=12, fontweight='bold')

plt.show()
```

# Churn Rate

```python
df.dtypes
```

```python
dft = df.copy()
```

```python
dft['InvoiceDate'] = pd.to_datetime(dft['InvoiceDate'])
dft['PurchaseMonth'] = dft['InvoiceDate'].dt.to_period('M')
```

```python
dft.dtypes
```

```python
# first purchase for each customer
first_purchase = dft.groupby('CustomerID').agg(
    FirstPurchaseMonth=('PurchaseMonth', 'min')
).reset_index()

first_purchase

# merging first purchase to dft to map all first purchases
dft = dft.merge(first_purchase, on='CustomerID', how='left')
```

```python
dft.head()
```

#### getting churn

```python
churn_threshold = 90  #days

def customer_type(row):
    if row['PurchaseMonth'] == row['FirstPurchaseMonth']:
        return "New Customer"
    if (dft['InvoiceDate'].max() - row['InvoiceDate']).days > churn_threshold:
        return "Churned Customer"
    return "Returning Customer"
```

```python
dft['Customer_Type'] = dft.apply(customer_type, axis=1)
```

```python
dft[['CustomerID', 'InvoiceDate', 'PurchaseMonth', 'Customer_Type']].head()
```

```python
customer_trends = dft.groupby(['PurchaseMonth', 'Customer_Type']).agg(
    CustomerCount=('CustomerID', 'nunique')
).reset_index()

customer_trends
```

```python
customer_trends.dtypes
```

```python
# customer_trends['CustomerCount'] = customer_trends['CustomerCount'].astype(int)
```

```python
# customer_trends['Customer_Type'] = customer_trends['Customer_Type'].astype(str)
```

```python
customer_trends['PurchaseMonth'] = customer_trends['PurchaseMonth'].dt.to_timestamp()
```

```python
plt.figure(figsize=(10, 6))

sns.lineplot(
    data=customer_trends,
    x='PurchaseMonth',
    y='CustomerCount',
    hue='Customer_Type',
    marker='o'
)

plt.xticks(rotation=45)
plt.xlabel('Purchase Month', fontsize=12, fontweight='bold')
plt.ylabel('Number of Customers', fontsize=12, fontweight='bold')
plt.title('Customer Retention & Churn Trends', fontsize=14, fontweight='bold', color='darkblue')
plt.legend(title='Customer Type')
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```
